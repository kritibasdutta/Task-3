# Inference from the dataset

Highest profit earned from the state of California where sales were also highest among all the states.
Highest loss or lowest profit has been registered in the state of Texas though the sales were 3rd highest.
Arizona, Colorado, Florida, Illinois, North Carolina, Ohio and Pennsylvania are among the loss making states.

Under categories, Furniture was making very less profit as compared to other categories

Under sub-categories, Table is making huge loss. The other loss making sub categories were - Bookcases and Supplies

In central region, consumer segment is incurring very less profit.

In south region, 'same day' ship mode is having losses.
